```python
#sepearting features and targets
x = data.drop('Exited', axis=1)
y = data['Exited']
```

```python
#next we perform train-test split 
from sklearn.model_selection import train_test_split

# Split into 80% train, 20% test
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

```

```python
!pip install imbalanced-learn
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

logreg = LogisticRegression(
    class_weight='balanced',
    max_iter=1000,
    random_state=42
)
logreg.fit(x_train, y_train)

rf = RandomForestClassifier(
    class_weight='balanced',
    n_estimators=100,
    random_state=42
)
rf.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Scale features (Logistic Regression needs this)
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Train Logistic Regression
logreg = LogisticRegression(
    class_weight='balanced',
    max_iter=2000,
    random_state=42
)

logreg.fit(x_train_scaled, y_train)
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    class_weight='balanced',
    n_estimators=200,
    random_state=42
)

rf.fit(x_train, y_train)
```

```python
#predicting probabilities
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Logistic Regression
y_pred_logreg = logreg.predict_proba(x_test_scaled)[:, 1]

# Random Forest
y_pred_rf = rf.predict_proba(x_test)[:, 1]
```

```python
#ROC-AUC scores
auc_logreg = roc_auc_score(y_test, y_pred_logreg)
auc_rf = roc_auc_score(y_test, y_pred_rf)

print(f"Logistic Regression ROC-AUC: {auc_logreg:.3f}")
print(f"Random Forest ROC-AUC: {auc_rf:.3f}")
```

```python
from sklearn.metrics import roc_auc_score

# Calculate AUC scores
auc_logreg = roc_auc_score(y_test, y_pred_logreg)
auc_rf = roc_auc_score(y_test, y_pred_rf)

print(f"Logistic Regression ROC-AUC: {auc_logreg:.3f}")
print(f"Random Forest ROC-AUC: {auc_rf:.3f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

fpr_log, tpr_log, _ = roc_curve(y_test, y_pred_logreg)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_rf)

plt.plot(fpr_log, tpr_log, label=f'LogReg (AUC={auc_logreg:.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'RF (AUC={auc_rf:.3f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
